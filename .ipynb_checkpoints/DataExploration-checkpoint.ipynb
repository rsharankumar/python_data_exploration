```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from textblob import TextBlob
import matplotlib as mpl
%matplotlib inline
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
from scipy import cluster
from scipy import stats
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
# Reading dataset
data = pd.read_csv('AS_NPS.csv', encoding = "ISO-8859-1")
# Printing the top 3 rows
data.head(3)
```

```python
data = data.fillna('')
# Remove spaces in column names
data.columns = data.columns.str.replace('\s+', '_')
# To replace [n/a] with null
data = data.replace('[n/a]', '')
# write to csv in utf format
data.to_csv('sample.csv', sep=',', encoding="utf-8") 
# read back the csv in utf encoding format
df = pd.read_csv('sample.csv')
# renaming columns
df = df.rename(columns={ df.columns[0]: "s_num" })
df = df.rename(columns={ df.columns[3]: "Account_Status" })
df = df.rename(columns={ df.columns[14]: "AustralianSuperMainFund" })
# print the top rows
df.head()
```

```python
df.Years_In_Fund_Bands = df.Years_In_Fund_Bands.replace('0 to 6 months', 'A_0 to 6 months')
df.Years_In_Fund_Bands = df.Years_In_Fund_Bands.replace('More than 6 months up to 1 year', 'B_More than 6 months up to 1 year')
df.Years_In_Fund_Bands = df.Years_In_Fund_Bands.replace('More than 1 year up to 5 years', 'C_More than 1 year up to 5 years')
df.Years_In_Fund_Bands = df.Years_In_Fund_Bands.replace('More than 5 years up to 10 years', 'D_More than 5 years up to 10 years')
df.Years_In_Fund_Bands = df.Years_In_Fund_Bands.replace('More than 10 years', 'E_More than 10 years')
###
df.Balance_Bands = df.Balance_Bands.replace('$0-10K', 'A_$0-10K')
df.Balance_Bands = df.Balance_Bands.replace('$>10K-20K', 'B_$>10K-20K')
df.Balance_Bands = df.Balance_Bands.replace('$>20K-50K', 'C_$>20K-50K')
df.Balance_Bands = df.Balance_Bands.replace('$>50K-100K', 'D_$>50K-100K')
df.Balance_Bands = df.Balance_Bands.replace('$>100K-200K', 'E_$>100K-200K')
df.Balance_Bands = df.Balance_Bands.replace('$>200K-500K', 'F_$>200K-500K')
df.Balance_Bands = df.Balance_Bands.replace('$>500K-1Mill', 'G_$>500K-1Mill')
df.Balance_Bands = df.Balance_Bands.replace('$>1Mill-2Mill', 'H_$>1Mill-2Mill')
df.Balance_Bands = df.Balance_Bands.replace('$>2Mill+', 'I_$>2Mill+')
###
df.Age = df.Age.replace('Under 18', 'A_Under 18')
df.Age = df.Age.replace('18-29', 'B_18-29')
df.Age = df.Age.replace('30-49', 'C_30-49')
df.Age = df.Age.replace('50-65', 'D_50-65')
df.Age = df.Age.replace('65+', 'E_65+')
```

```python


#ps["comment"] = ps["Comment"].map(str)
def get_polarity(text):
    textblob = TextBlob(text)
    pol = textblob.sentiment.polarity
    return round(pol,2)

def get_subjectivity(text):
    textblob = TextBlob(text)
    subj = textblob.sentiment.subjectivity
    return round(subj,2)
```

```python
ps_comp = df[['Comment','NPS_Competitor_Fund']]
ps_comp = ps_comp.dropna() 
ps_comp['polarity'] = ps_comp['Comment'].apply(get_polarity)
ps_comp['subjectivity'] = ps_comp['Comment'].apply(get_subjectivity)
print('Correlation between Competition NPS and Polarity:', ps_comp['NPS_Competitor_Fund'].corr(ps_comp['polarity']))
print('Correlation between Competition NPS and Subjectivity:', ps_comp['NPS_Competitor_Fund'].corr(ps_comp['subjectivity']))
```

```python
ps_as = df[['Comment.1','AustralianSuperMainFund']]
ps_as = ps_as.dropna() 
ps_as['polarity'] = ps_as['Comment.1'].apply(get_polarity)
ps_as['subjectivity'] = ps_as['Comment.1'].apply(get_subjectivity)
print('Correlation between Competition NPS and Polarity:', ps_as['AustralianSuperMainFund'].corr(ps_as['polarity']))
print('Correlation between Competition NPS and Subjectivity:', ps_as['AustralianSuperMainFund'].corr(ps_as['subjectivity']))
```

```python
df_tr = df_tr.dropna()
```

```python
#Transsform the timeOfDay to dummies
df_tr = pd.get_dummies(df_tr, columns=['Geolocation'])
```

```python
print('Number of Rows', df_tr.shape[0])
df_tr.head(3)
```

```python
#Standardize
clmns = ['Years_In_Fund_Bands', 'Balance_Bands', 'Age', 'AustralianSuperMainFund', 'Geolocation_ACT', 'Geolocation_NSW', 'Geolocation_NT', 'Geolocation_QLD', 'Geolocation_SA', 'Geolocation_TAS', 'Geolocation_VIC', 'Geolocation_WA']
df_tr_std = stats.zscore(df_tr[clmns])
df_tr_std = df_tr
```

```python
# To Identofy Ideal Clusters
#plot variance for each value for 'k' between 1,10
initial = [cluster.vq.kmeans(df_tr_std,i) for i in range(1,10)]
plt.plot([var for (cent,var) in initial])
plt.show()
```

```python
#Cluster the data
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_tr_std)
labels = kmeans.labels_
```

```python
#Glue back to originaal data
df_tr['clusters'] = labels
```

```python
#Add the column into our list
clmns.extend(['clusters'])
```

```python
#Lets analyze the clusters
print(df_tr[clmns].groupby(['clusters']).mean())
```
